```python
from google.colab import drive
drive.mount('/content/drive')
csv_directory = '/content/drive/My Drive/ankit_data_vis/'
import os
import pandas as pd

# List files in the directory
csv_files = os.listdir(csv_directory)

# Load CSV files into DataFrames
dfs = {}
for file in csv_files:
    if file.endswith('.csv'):
        df_name = os.path.splitext(file)[0]  # Use file name as DataFrame name
        df = pd.read_csv(os.path.join(csv_directory, file))
        dfs[df_name] = df
# Access the DataFrame for 'poppy.csv'
poppy_df = dfs['poppy']
nitrous_df = dfs['nitrous']
desoxypipradol_df = dfs['desoxypipradol']
loperamide_df = dfs['loperamide']
```
